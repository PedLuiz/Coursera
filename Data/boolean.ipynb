```python
import pandas as pd
import numpy as np
```

```python
data = {'planet': ['Mercury', 'Venus', 'Earth', 'Mars', 
                    'Jupiter', 'Saturn', 'Uranus', 'Neptune'], 
        'radius_km': [2440, 6052, 6371, 3390, 66911, 58232, 25362, 24622],
        'moons': [0, 0, 1, 2, 80, 83, 27, 14],
        'type': ['terrestrial', 'terrestrial',         
                'terrestrial', 'terrestrial', 'gas giant', 'gas giant', 'ice giant', 'ice giant'],
        'rings': ['no', 'no', 'no', 'no', 'yes', 'yes',
                'yes', 'yes'],
        'mean_temp_c': [167, 464, 15, -65, -110, -140, -195, -200],
        'magnetic_fields': ['yes', 'no', 'yes', 'no', 'yes', 'yes', 'yes', 'yes']
        }
planets = pd.DataFrame(data)
planets
```

```python
mask = planets['moons'] < 20
mask
```

```python
planets[mask]
```

```python
mask = (planets['moons'] < 10) | (planets['moons'] > 50)
planets[mask]
```

```python
mask = (planets['moons'] > 20) & (planets["moons"] != 80) & (planets['radius_km'] >= 50000)
planets[mask]
```

```python
planets.head()
```

```python
# groupby faz um DataFrame agrupando uma coluna especificada
planets.groupby(["type"]).sum(numeric_only=True)[['moons']]
```

```python
planets.groupby(['type', 'magnetic_fields']).mean(numeric_only=True)
```

```python
# agg() permite aplicar formulas em grupos
planets.groupby(['type', 'magnetic_fields'])[planets.select_dtypes(include='number').columns].agg(['mean', 'max'])
```

```python
planets.groupby(['type', 'magnetic_fields']).max()
```
